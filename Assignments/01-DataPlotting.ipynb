```
# Importing Packages
import matplotlib.pyplot as plt
import statistics
import random
```

```
# Generating Random Data
roll = [f'{i}' for i in range(1, 51)]
marks = [random.randint(0,20) for i in roll]
```

```
# Calculating mean, median, mode and standard deviation
mean = statistics.mean(marks)
median = statistics.median(marks)
stdev = statistics.stdev(marks)
mode = statistics.mode(marks)
```

```
# Plotting Data
fig, ax = plt.subplots(figsize=(14, 5), facecolor='w')
ax.plot(roll, [mean for i in roll], color='m', label=f"Mean: {round(mean, 2)}")
ax.plot(roll, [median for i in roll], color='r',  label=f"Median: {round(median, 2)}")
ax.plot(roll, [mode for i in roll], color='g', label=f"Mode: {round(mode, 2)}")
ax.plot(roll, [stdev for i in roll], color='k', label=f"Std Dev: {round(stdev, 2)}")
ax.bar(roll, marks, label="Marks", color='c')
ax.set_xlabel("Roll No")
ax.set_ylabel("Mark")
ax.set_title("Marks of Random Students (Out of 20)")
ax.legend()
plt.show()
```
